Step 1: Load SpaceX Launch Data from API

```python
#Use https://api.spacexdata.com/v4/launches
import requests
url = 'https://api.spacexdata.com/v4/launches'
response = requests.get(url)
response
```

```python
launch = response.json()
launch = response.json()
launch = pd.DataFrame(launch)
launch.head()
```

```python
#Extract relevant columns: name, date_utc, success, details, rocket
df_launch=launch[['name', 'date_utc', 'success', 'details', 'rocket']]
```

```python
#Convert date_utc to datetime and extract the year
df_launch['date_utc'] = pd.to_datetime(df_launch['date_utc'])
df_launch['year'] = df_launch['date_utc'].dt.year
df_launch.head()
```

Step 2: Load Rocket Metadata

```python
#Use  https://api.spacexdata.com/v4/rockets
import requests
url_rocket = 'https://api.spacexdata.com/v4/rockets'
response = requests.get(url_rocket)
response
```

```python
#Extract id, name, type, active, and stages

df_rocket = response.json()
df_rocket = pd.DataFrame(df_rocket)
df_rocket.head()
```

```python
#Extract id, name, type, active, and stages
df_rocket = df_rocket[['id','name','type','active','stages']]
df_rocket.head()
```

Step 3: Merge Launch and Rocket Data

```python
#Join the two DataFrames on rocket ID using
df_merged = pd.merge(df_launch, df_rocket, left_on='rocket', right_on='id', how='left')
df_merged = df_merged.drop('id', axis=1)
df_merged.head()
```

Step 4: Add Simulated Country Information

```python
#Add a new column country and randomly assign one of these values: ['USA', 'Russia', 'India', 'China', 'France']
import random
df_merged['country'] = [random.choice(['USA', 'Russia', 'India', 'China', 'France']) for _ in range(len(df_merged))]
df_merged.head()
```

Step 5: Store Merged Data in SQLite3

```python
#Use sqlite3 to create a connection and save the merged DataFrame as a table named launches
#Table should contain all merged columns including country

import sqlite3

# Create a connection to a new SQLite database file (or in-memory)
conn = sqlite3.connect('spacex_launches.db')  # This creates a file named spacex_launches.db

# Save the DataFrame as a table named 'launches'
df_merged.to_sql('launches', conn, if_exists='replace', index=False)

conn.close()
```

Step 6: Run SQL Queries on the Data to analyze



```python
#connected to database
conn = sqlite3.connect('spacex_launches.db')
```

```python
#displaying the table
q1 = "SELECT * FROM launches LIMIT 5"
pd.read_sql(q1, conn)
```

```python
#Launches by Country
q2 ="SELECT country, COUNT(*) AS Launches FROM launches GROUP BY country"
pd.read_sql(q2,conn)
```

```python
#Which year had the highest number of launches?
q3 ="SELECT year, COUNT(*) AS count FROM launches GROUP BY year ORDER BY count DESC LIMIT 1"
pd.read_sql(q3,conn)
```

```python
#Top 5 Missions by Launch Count
q5 = "SELECT name_x, COUNT(*) AS count FROM launches GROUP BY name_x ORDER BY count DESC LIMIT 5"
pd.read_sql(q5,conn)
```
